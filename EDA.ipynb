```python
import os
```

```python


def count_images_in_folders(root_folder):
    # Get the list of subdirectories (class labels)
    class_labels = os.listdir(root_folder)

    # Initialize a dictionary to store counts for each class
    class_counts = {}

    # Loop through each class label (subdirectory)
    for label in class_labels:
        # Construct the full path to the class folder
        class_folder = os.path.join(root_folder, label)

        # Check if it's a directory
        if os.path.isdir(class_folder):
            # Get the list of files in the folder
            files = os.listdir(class_folder)

            # Count the number of files (images) in the folder
            num_images = len([file for file in files if file.lower().endswith(('.png', '.jpg', '.jpeg','webp'))])

            # Store the count in the dictionary
            class_counts[label] = num_images

    return class_counts

```

```python

# Specify the root folder containing subfolders for each class
root_folder = 'testset'  # Replace with the actual path to your dataset

# Get the counts
test_image_counts = count_images_in_folders(root_folder)

# Display the counts
for label, count in test_image_counts.items():
    print(f"Number of images in {label}: {count}")
```

```python
# Specify the root folder containing subfolders for each class
root_folder = 'data'  # Replace with the actual path to your dataset

# Get the counts
train_image_counts = count_images_in_folders(root_folder)

# Display the counts
for label, count in train_image_counts.items():
    print(f"Number of images in {label}: {count}")
```

```python
distribution_of_dataset = {}
list_label = ["Happy","Angry","Neutral"]
for x in list_label:
    distribution_of_dataset[x] = test_image_counts[x] + train_image_counts[x]

distribution_of_dataset
```

```python
import matplotlib.pyplot as plt

def plot_bar_graph(class_counts):
    labels = list(class_counts.keys())
    counts = list(class_counts.values())

     # Add data labels to each bar
    for i, count in enumerate(counts):
        plt.text(i, count + 1, str(count), ha='center', va='bottom', fontweight='bold')

    plt.bar(labels, counts, color=['blue', 'red', 'green'])
    plt.xlabel('Emotion Classes')
    plt.ylabel('Number of Images')
    plt.title('Number of Images in Each Emotion Class')
    plt.show()
```

```python
def plot_pie_chart(class_counts):
    labels = list(class_counts.keys())
    counts = list(class_counts.values())

    plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=['blue', 'red', 'green'], startangle=90)
    plt.title('Percentage of classes')
    plt.show()
```

```python
plot_bar_graph(distribution_of_dataset)
plot_pie_chart(distribution_of_dataset)
```
